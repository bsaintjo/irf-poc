```python
import pandas
```

```python
df = pandas.read_csv("../data/MIR_Fruit_purees.csv", header=None)
df.head()
```

### Preprocessing

* Remove the wavenumber column
* Convert strawbery to binary label
* Split into test train datasets

```python
column_names = {idx: str(wavenumber) for idx, wavenumber in enumerate(df.iloc[:, 0])}
column_names[0] = "is_strawberry"
print(column_names)
```

```python
strawberry_df = df.drop(columns=0).T.rename(column_names, axis="columns")
strawberry_df.head()
```

```python
strawberry_df.iloc[:, 0].replace({"NON-Strawberry": 0, "Strawberry": 1}, inplace=True)
strawberry_df.head()
```

```python
print(
    "Number of strawberry samples: ", strawberry_df[strawberry_df.iloc[:, 0] == 1].shape
)
print(
    "Number of non-strawberry samples: ",
    strawberry_df[strawberry_df.iloc[:, 0] == 0].shape,
)
```

```python
from sklearn.model_selection import train_test_split

y = strawberry_df.iloc[:, 0]
X = strawberry_df.iloc[:, 1:]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train.shape
```

```python
from irf import irf_utils
from irf.ensemble import RandomForestClassifierWithWeights

(
    all_rf_weights,
    all_K_iter_rf_data,
    all_rf_bootstrap_output,
    all_rit_bootstrap_output,
    stability_score,
) = irf_utils.run_iRF(
    X_train=X_train,
    X_test=X_test,
    y_train=y_train,
    y_test=y_test,
    K=5,  # number of iteration
    rf=RandomForestClassifierWithWeights(n_estimators=20),
    B=30,
    random_state_classifier=2018,  # random seed
    propn_n_samples=0.2,
    bin_class_type=1,
    M=20,
    max_depth=5,
    noisy_split=False,
    num_splits=2,
    n_estimators_bootstrap=5,
)
```
